```python
import pandas as pd
import numpy as np


df = pd.read_csv('/content/myexcel - myexcel.csv.csv')


df['height'] = np.random.randint(150, 181, size=len(df))


df.drop_duplicates(inplace=True)
df.fillna(method='ffill', inplace=True)
```

```python
print(df.columns)
```

```python
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / len(df)) * 100
```

```python
position_distribution = df['Position'].value_counts()
```

```python
age_bins = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60], labels=['20-30', '30-40', '40-50', '50-60'])
age_group_distribution = age_bins.value_counts()
```

```python
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().reset_index()
highest_expenditure = salary_expenditure.loc[salary_expenditure['Salary'].idxmax()]
```

```python
correlation = df['Age'].corr(df["Salary"])

import matplotlib.pyplot as plt
plt.scatter(df['Age'], df['Salary'])
plt.title('Correlation between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar Plot
team_distribution = df['Team'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values, palette='viridis')
plt.title('Distribution of Employees Across Teams', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Age Groups
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(6,6))
sns.histplot(df['Age'], bins=8, kde=True, color='skyblue')
plt.title('Age Distribution of Employees', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Stacked Bar Chart for Salary Expenditure
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum().unstack()

salary_expenditure.plot(kind='bar', stacked=True, figsize=(7, 7), colormap='viridis')
plt.title('Salary Expenditure by Team and Position', fontsize=16)
plt.xlabel('Teams', fontsize=12)
plt.ylabel('Total Salary Expenditure', fontsize=12)
plt.legend(title='Positions', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot for Age vs. Salary
plt.figure(figsize=(12, 10))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Team', palette='viridis', s=100, alpha=0.7)
plt.title('Correlation Between Age and Salary', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
